## 선행종합지수

* 각 국가별, 지역별로 3 ~ 10 (6 ~ 9)개월 후의 경기흐름을 예측하는 지수
* 국가,지역의 경기 전환점 예측을 위해 이용
* 통계청 경기 선행 지수[7개]
    * 재고순환지표, 코스피지수, 장단기금리차, 기계류내수출하지수(선박제외), 건설수주액, 경제심리지수, 수출입물가비율
* OECD 경기 선행 지수[6개]
    * 재고순환지표, 코스피 지수, 장단기 금리차(3년물-1일물 금리), 업황, 재고량,  순교역조건
* https://kostat.go.kr/understand/info/info_lge/1/detail_lang.action?bmode=detail_lang&pageNo=1&keyWord=7&cd=SL4428&sTt=
    * 통상 지표가 현재까지와 반대방향으로 2분기 이상 연속하여 움직이면, 이 시점(선행지수 순환변동치의 전환 시점)을 경기전환점 발생 신호로 보고, 여기에 과거의 평균 선행시차를 더하면 향후 국면전환이 발생할 시점을 대략 추정

```python
from datetime import datetime
import requests
import json
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import yfinance as yf

plt.rcParams["figure.figsize"] = (10, 5)
plt.rcParams['lines.linewidth'] = 1
plt.rcParams['lines.color'] = 'b'

# Composite Leading Indicator
# https://data.oecd.org/leadind/composite-leading-indicator-cli.htm
def getCLI(country):
    uri = 'https://stats.oecd.org/sdmx-json/data/DP_LIVE/' + country + '.CLI.AMPLITUD.LTRENDIDX.M/OECD?json-lang=en&dimensionAtObservation=allDimensions&startPeriod=2005-01&endPeriod=2020-12'
    resp = requests.get(uri)
    result = json.loads(resp.text)
    dates = []
    cli = []
    cli_code = 'CLI.' + country

    observations = result['dataSets'][0]['observations']
    for key in observations:
        obs = observations[key][0]
        cli.append(obs)


    time_period = result['structure']['dimensions']['observation'][5]['values']

    for date in time_period:
        date = date['id']
        year = int(date[:4])
        month = int(date[5:7])
        dates.append(datetime(year=year, month=month, day=1))
    df = pd.DataFrame(data=cli, index=dates, columns=['cli'])
    return df

def GetYahooFinance(name, code):
    ticker = yf.Ticker(code)
    ticker = ticker.history(period='16y')
    ticker = ticker[['Close']]
    ticker.rename(columns={'Close': name}, inplace=True)
    return ticker

code_names = [('KOR', 'Kospi', '^KS11'),
              ('USA', 'Nasdaq', '^DJI')]

df_list = []
for country_code, yf_name, yf_code in code_names:
    cli_code = 'CLI.' + country_code
    cli = getCLI(country_code)
    ticker = GetYahooFinance(yf_name, yf_code)
    df = pd.concat([cli['cli'], ticker[yf_name]], axis=1)
    df['country'] = country_code
    df.columns = ['cli','index','country']
    df = df.interpolate(limit_direction='backward')
    df_list.append(df)
    
df = pd.concat(df_list)    
# print(df.head())


fig = plt.figure()
ax = fig.add_subplot(111)
df[df['country']=='KOR']['cli'].dropna().plot(label="CLI.KOR")
ax2 = ax.twinx()
df[df['country']=='KOR']['index'].dropna().plot(label="kospi", linestyle='-',color='orangered')
fig.legend(loc="upper left", bbox_to_anchor=(0, 1), bbox_transform=ax.transAxes)
plt.show()

fig = plt.figure()
ax = fig.add_subplot(111)
df[df['country']=='USA']['cli'].dropna().plot(label="CLI.USA")
ax2 = ax.twinx()
df[df['country']=='USA']['index'].dropna().plot(label="nasdaq", linestyle='-',color='orangered')
fig.legend(loc="upper left", bbox_to_anchor=(0, 1), bbox_transform=ax.transAxes)
plt.show()
```
