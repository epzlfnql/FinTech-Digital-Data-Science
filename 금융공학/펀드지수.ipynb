```python
일자별주가.set_index('일자',inplace=True)
for i in 일자별주가.columns:
    일자별주가[f'{i} 시총가'] = 일자별주가[f'{i}']* 종목기본정보[종목기본정보.종목명==i]['발행주식수'].values 
일자별주가['시가총액합계'] = 일자별주가['A전자 시총가'] + 일자별주가['B건설 시총가']+ 일자별주가['C증권 시총가'] + 일자별주가['D화학 시총가'] + 일자별주가['E엔터 시총가']
```

```python
일자별주가['지수'] = 일자별주가['시가총액합계'] / 일자별주가['시가총액합계'].iloc[0]*100
```

```python
# 시가총액 편입비중 구하기~
일자별주가['A전자_편입비중'] = 일자별주가['A전자 시총가'] / 일자별주가['시가총액합계']
일자별주가['B건설_편입비중'] = 일자별주가['B건설 시총가'] / 일자별주가['시가총액합계']
일자별주가['C증권_편입비중'] = 일자별주가['시가총액합계'] / 일자별주가['시가총액합계']
일자별주가['D화학_편입비중'] = 일자별주가['D화학 시총가'] / 일자별주가['시가총액합계']
일자별주가['E엔터_편입비중'] = 일자별주가['E엔터 시총가'] / 일자별주가['시가총액합계']
일자별주가
```

```python
# 설정좌수랑 환매좌수, 일자별 매매내역은 그냥 갖고오기
펀드 = pd.read_excel('./data/lec04_펀드_자산 - 복사본.xlsx', sheet_name = '펀드')
펀드.columns = 펀드.iloc[0]
펀드.drop(펀드.index[0], inplace=True)

tmp = 펀드.iloc[:,12:17]
tmp.columns = ['A전자 일자별 매매내역', 'B건설 일자별 매매내역', 'C증권 일자별 매매내역', 'D화학 일자별 매매내역', 'E엔터 일자별 매매내역']
일자별주가['설정좌수'] = 펀드['설정좌수']
일자별주가['환매좌수'] = 펀드['환매좌수']





# 총좌수 구하기
일자별주가['총좌수'] = 일자별주가['설정좌수'] - 일자별주가['환매좌수']
일자별주가['총좌수'] = 일자별주가['총좌수'].cumsum()
일자별주가['iNAV'] = 10000*일자별주가['지수'] / 일자별주가['지수'].iloc[0]
일자별주가['AUM'] = 일자별주가['iNAV'] * 일자별주가['총좌수']
일자별주가 = pd.concat([일자별주가, tmp ],axis=1)

# 주식잔고 cumsum해버리기
일자별주가['A전자 주식잔고'] = 일자별주가['A전자 일자별 매매내역'].cumsum()
일자별주가['B건설 주식잔고'] = 일자별주가['B건설 일자별 매매내역'].cumsum()
일자별주가['C증권 주식잔고'] = 일자별주가['C증권 일자별 매매내역'].cumsum()
일자별주가['D화학 주식잔고'] = 일자별주가['D화학 일자별 매매내역'].cumsum()
일자별주가['E엔터 주식잔고'] = 일자별주가['E엔터 일자별 매매내역'].cumsum()


일자별주가
```
